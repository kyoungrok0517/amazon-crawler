```python
import pandas as pd
import json
import pymongo
from pymongo import MongoClient
import codecs
from pprint import pprint
```

```python
client = MongoClient()
db = client['amazon']
```

# DataFrame for Items

```python
df_items = pd.DataFrame(list(db.headphone_detail.find(projection={'_id': False, 'item_type': False, 'images': False})))
df_items.set_index(['title', 'link'])
```

```python
df_items.count()
```

# DataFrame for Reviews

```python
df_reviews = pd.DataFrame(list(db.headphone_review.find(projection={'_id': False, 'item_type': False})))
df_reviews.set_index(['title', 'item_link'])
```

```python
join_text = lambda x: "\n".join(x)
joined = df_reviews['text'].map(join_text)
df_reviews.update(joined)
```

```python
df_reviews.count()
```

# Join Items & Reviews

```python
df_merged = pd.merge(left=df_items, right=df_reviews, left_on='link', right_on='item_link', sort=True, suffixes=('_item', '_review'))
```

```python
df_merged.dtypes
```

```python
df_merged.get(1)
```

```python
df_merged.set_index(['title_item'])
df_merged.sort(['review_count'], ascending=False)
```

# Save the Result

```python
collection = db['headphone_merged']

collection.insert_many(df_merged.to_dict('record'))
```

```python
collection.create_index([("title_item", pymongo.ASCENDING), ("helpful_vote_count", pymongo.DESCENDING)])
```

```python
pprint(list(collection.find(limit=3, sort=[('helpful_vote_count', -1)], projection=('title_item', 'text', 'features', 'helpful_vote_count'))))
```
